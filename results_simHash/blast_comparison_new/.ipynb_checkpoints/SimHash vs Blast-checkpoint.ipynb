# Initialise libs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from decimal import Decimal
import numpy as np
```

# Load the results

```python
ENSP233_blash4 = pd.read_csv('ENSP233/ENSP233_blash_4_3_compare.csv')
ENSP233_blash8 = pd.read_csv('ENSP233/ENSP233_blash_8_3_compare.csv')
ENSP233_blash12 = pd.read_csv('ENSP233/ENSP233_blash_12_3_compare.csv')
ENSP233_blash16 = pd.read_csv('ENSP233/ENSP233_blash_16_3_compare.csv')
ENSP233_blash20 = pd.read_csv('ENSP233/ENSP233_blash_20_3_compare.csv')


ENSP233_native = pd.read_csv('ENSP233/ENSP233_native_3_compare.csv')
ENSP233_spooky = pd.read_csv('ENSP233/ENSP233_spooky_3_compare.csv')

ENSP1146_blash12 = pd.read_csv('ENSP1146/ENSP1146_blash_12_3_compare.csv')
ENSP1146_native = pd.read_csv('ENSP1146/ENSP1146_native_3_compare.csv')
ENSP1146_spooky = pd.read_csv('ENSP1146/ENSP1146_spooky_3_compare.csv')
```

# Spooky comparison (window length 3)

```python
plt.figure(figsize=(15,8))
sns.regplot(x="bitscore", y="similarity",data=ENSP233_spooky)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Corelation - Window size 3, Spooky for ENSP233(bitscore)", fontsize = 32)
#plt.savefig("SimHash Error vs Blast corelation - Window size 3, boost for ENSP233(bitscore).png")

plt.figure(figsize=(15,8))
sns.regplot(x="bitscore", y="similarity",data=ENSP1146_spooky)
plt.xlabel('Bsitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Corelation - Window size 3, Spooky for ENSP1146(bitscore)", fontsize = 32)
#plt.savefig("SimHash Error vs Blast corelation - Window size 3, boost for ENSP1146(bitscore).png")
```

# BLASH performance on ENSP233

```python
plt.figure(figsize=(15,8))
sns.regplot(x="bitscore", y="similarity",data=ENSP233_blash4)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Corelation - Window size 3, Blash_4 for ENSP233(bitscore)", fontsize = 32)

plt.figure(figsize=(15,8))
sns.regplot(x="bitscore", y="similarity",data=ENSP233_blash8)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Corelation - Window size 3, Blash_8 for ENSP233(bitscore)", fontsize = 32)

plt.figure(figsize=(15,8))
sns.regplot(x="bitscore", y="similarity",data=ENSP233_blash12)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Corelation - Window size 3, Blash_12 for ENSP233(bitscore)", fontsize = 32)
```

```python
plt.figure(figsize=(15,8))
sns.regplot(x="bitscore", y="similarity",data=ENSP233_blash16)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Corelation - Window size 3, Blash_16 for ENSP233(bitscore)", fontsize = 32)

plt.figure(figsize=(15,8))
sns.regplot(x="bitscore", y="similarity",data=ENSP233_blash20)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Corelation - Window size 3, Blash_20 forENSP233(bitscore)", fontsize = 32)
```

```python
blash_comparison = [ENSP233_native, ENSP233_blash4, ENSP233_blash8, ENSP233_blash12, ENSP233_blash16,
                    ENSP233_blash20]

blash_comparison_frame = pd.concat(blash_comparison)
```

```python
plt.figure(figsize=(15,8))
sns.lmplot(x="bitscore", y="similarity", hue="hash", size=14, aspect=1.5, data=blash_comparison_frame,
           legend_out=False)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Blash comparison", fontsize = 32)
plt.legend(loc='upper left',fontsize=20)
```

```python
best_blash = [ENSP233_native,  ENSP233_blash12, ENSP233_blash16, ENSP233_blash20]
best_blash_frame = pd.concat(best_blash)
```

```python
plt.figure(figsize=(15,8))
sns.lmplot(x="bitscore", y="similarity", hue="hash", size=14, aspect=1.5, data=best_blash_frame, legend_out=False)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Blash comparison", fontsize = 32)
plt.legend(loc='upper left',fontsize=20)
```

```python
best_blash_2 = [ENSP233_native, ENSP233_blash16, ENSP233_blash20]
best_blash_frame_2 = pd.concat(best_blash_2)

plt.figure(figsize=(15,8))
sns.lmplot(x="bitscore", y="similarity", hue="hash", size=14, aspect=1.5, data=best_blash_frame_2, legend_out=False)
plt.xlabel('Bitscore - BLAST', fontsize=18)
plt.ylabel('Similarity', fontsize=18)
plt.suptitle("Blash comparison", fontsize = 32)
plt.legend(loc='upper left',fontsize=20)
```
